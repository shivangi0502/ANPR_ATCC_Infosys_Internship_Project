```python

import os
import yaml
from ultralytics import YOLO


OLD_YAML_PATH = r"C:\Users\ADMIN\Documents\Dev_Projects\Traffic_Infosys\uvh26_final.yaml"


NEW_DATA_ROOT = r"C:\Users\ADMIN\Documents\Dev_Projects\Traffic_Infosys\new_finetunning\finetune_dataset"
TRAIN_FOLDER = "train" # Folder name inside root
VAL_FOLDER = "valid"   # Folder name inside root


MODEL_PATH = r"C:\Users\ADMIN\Documents\Dev_Projects\Traffic_Infosys\UVH_Traffic_Project\yolov9e_a6000_run\weights\best.pt"
```

```python
print(f"Reading class names from: {OLD_YAML_PATH}")
with open(OLD_YAML_PATH, 'r') as f:
    config = yaml.safe_load(f)
```

```python
config['path'] = NEW_DATA_ROOT
config['train'] = TRAIN_FOLDER
config['val'] = VAL_FOLDER
```

```python
new_yaml_path = os.path.join(NEW_DATA_ROOT, "finetune_run.yaml")
print(f"Creating temporary training config: {new_yaml_path}")

with open(new_yaml_path, 'w') as f:
    yaml.dump(config, f, default_flow_style=False)
```

```python
print("Loading model...")
model = YOLO(MODEL_PATH)

print("Starting Fine-Tuning on new dataset...")
results = model.train(
    data=new_yaml_path, 
    epochs=50,         
    imgsz=640,
    batch=8,            
    lr0=0.001,        
    lrf=0.01,
    freeze=10,         
    project="FineTune_Project",
    name="scooter_fix_v1"
)

print("Training Complete!")
```
