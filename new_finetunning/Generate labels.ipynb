```python
import os
from ultralytics import YOLO


MODEL_PATH = r"C:\Users\ADMIN\Documents\Dev_Projects\Traffic_Infosys\UVH_Traffic_Project\yolov9e_a6000_run\weights\best.pt"

IMAGES_DIR = r"C:\Users\ADMIN\Documents\Dev_Projects\Traffic_Infosys\new_finetunning\finetune_dataset\images"

LABELS_DIR = r"C:\Users\ADMIN\Documents\Dev_Projects\Traffic_Infosys\new_finetunning\finetune_dataset\labels"

os.makedirs(LABELS_DIR, exist_ok=True)

print(f"Loading model from: {MODEL_PATH}")
model = YOLO(MODEL_PATH)

print(f"Processing images from: {IMAGES_DIR}...")


results = model.predict(source=IMAGES_DIR, conf=0.25, stream=True)

count = 0

for result in results:
    # Get the original image filename (e.g., 'frame_0.jpg')
    image_name = os.path.basename(result.path)
    # Change extension to .txt (e.g., 'frame_0.txt')
    txt_name = os.path.splitext(image_name)[0] + ".txt"
    # Full path to save the .txt file
    save_path = os.path.join(LABELS_DIR, txt_name)
    
    # Get detections: Class IDs and Normalized Coordinates (x, y, w, h)
    boxes = result.boxes.xywhn  # Normalized xywh
    classes = result.boxes.cls  # Class IDs
    
    with open(save_path, "w") as f:
        for cls, box in zip(classes, boxes):
            # YOLO Format: class_id x_center y_center width height
            c = int(cls)
            x, y, w, h = box.tolist()
            
            # Write line to file
            f.write(f"{c} {x:.6f} {y:.6f} {w:.6f} {h:.6f}\n")
    
    count += 1

print(f"Success! Generated labels for {count} images.")
print(f"Labels saved to: {LABELS_DIR}")
```
